```python
from scipy.stats import norm
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#Brownian Motion Process
#X(0) = X0 & X(t + dt) = X(t) + N(0, sigma**2 * dt)

x = 50
sigma = 0.5
totalTime = 1000
steps = 200
dt = totalTime/steps

for i in range(10):
    BrownianMotionProcess = []
    time = []
    for j in range(steps):
        x += norm.rvs(scale = sigma**2 * dt)
        BrownianMotionProcess.append(x)
    t = 0
    for k in range(steps):
        t += dt
        time.append(t)
    plt.plot(time, BrownianMotionProcess)
```
