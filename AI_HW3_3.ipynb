```python
from scipy.stats import norm
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
sigma_x = 0.5
sigma_y = 0.8
totalTime = 100000
dt = 5
steps = int(totalTime/dt) #in this case, 20,000 stepsÔºÅ

x = 50
y = 50
BrownianMotionProcess_x = []
BrownianMotionProcess_y = []
for j in range(steps):
    x += norm.rvs(scale = sigma_x**2 * dt)
    y += norm.rvs(scale = sigma_y**2 * dt)
    BrownianMotionProcess_x.append(x)
    BrownianMotionProcess_y.append(y)
plt.plot(BrownianMotionProcess_x, BrownianMotionProcess_y)
print("This process starts at (x = {},y = {}), and it ends at (x = {},y = {}).".format(50, 50, BrownianMotionProcess_x[-1], BrownianMotionProcess_y[-1]))
```
